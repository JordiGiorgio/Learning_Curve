```python
# Import the required libraries
from PIL import Image
import pandas as pd, numpy as np, seaborn as sns, matplotlib.pyplot as plt
```

```python
# Load the data
dogs = Image.open(r'C:\Users\Jordi\OneDrive\Documents\EDUCATION\SIMPLILEARN\Artificial Intelligence Master Program\MACHINE LEARNING\SELF LEARNING\Lesson 7\Datasets\dogs.jpeg')
```

```python
# View the data
#dogs.rotate(45).show(dogs)
plt.imshow(dogs)
```

```python
# Convert the image into ndarray
dogs_array = np.asarray(dogs)
print(type(dogs_array))
print(dogs_array.shape)
```

```python
# Convert the image in 2D array
dogs_array_new = dogs_array.reshape(185*272,3)
print(dogs_array_new.shape)
dogs_array_new
```

    Use K-means clustering with k set to 3 and cluster the image :

```python
# Import KMeans
from sklearn.cluster import KMeans

# Instantiate the function
kmeans = KMeans(n_clusters=3)

# Fit the data
kmeans.fit(dogs_array_new)
```

    Predict the cluster label of every pixel in the image and plot it back as an image :

```python
unique, counts = np.unique(kmeans.labels_, return_counts=True)
print(dict(zip(unique, counts)))
```

```python
# Predictions
dogs_pred = kmeans.predict(dogs_array_new)

# Reshape the array
dogs_pred2 = dogs_pred.reshape(185,272)
```

```python
plt.imshow(dogs_pred2)
```

    Find out the three dominant colors in the image :

```python
centers = [centers.astype(int) for centers in kmeans.cluster_centers_]
```
